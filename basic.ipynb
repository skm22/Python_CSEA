```python
def even_Fact(n):
    ans=1
    for i in range(2,n+1,2):
        ans*=i
    return ans

def odd_Fact(n):
    ans=1
    for i in range(1,n+1,2):
        ans*=i
    return ans

n=int(input("ENter the no"))
if n%2==0:
    print(even_Fact(n))
else:
    print(odd_Fact(n))
```
