Q3. W.A.P to Determine grade based on marks.
marks >= 90: Grade A
marks >= 80: Grade: B
Marks >= 70: Grade C
marks >= 60: Grade: D
otherwise,
Grade: E

```python
grade=int(input("Enter the grade"))
if grade>=90:
    print("Grade A")
elif grade>=80:
    print("Grade B")
elif grade>=70:
    print("Grade C")
elif grade >=60:
    print("Grade D")
else:
    print("Fail")
```

Q19. Write a program to input electricity unit charges and calculate
total electricity bill
according to the given condition:
• For first 50 units Rs. 0.50/unit
• For next 150 units Rs. 0.75/unit
• For next 250 units Rs. 1.20/unit
• For unit above 450 Rs. 1.50/unit
• An additional surcharge of 20% is added to the bill

```python
u = int(input('ENter the no of units consumed'))
if u<=50:
    bill=u*0.50
elif  u<=200:
    bill =50*0.50+(u-50)*0.75
elif u<=450:
    bill=50*0.50+150*0.75+(u-200)*1.20
else:
    bill=50*0.50+150*0.75+250*1.20+(u-450)*1.5

final_bill=bill+bill*0.2
print("Final bill is ",final_bill)
```
